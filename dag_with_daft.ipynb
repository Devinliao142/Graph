```python
import daft
import matplotlib
%matplotlib inline

from matplotlib import rc
rc("font", family="serif", size=12)
rc("text", usetex=True)
rc("./weaklensing.tex")
```

```python
# customize color
p_color = {"ec": "#46a546"}
s_color = {"ec": "#f89406"}
```

```python
pgm = daft.PGM([5., 2.6], origin=[-1.52, 2.], directed=False)
pgm.add_node(daft.Node("IP1", '(40.7,74.0)', 1.5, 4, aspect=2, scale=1.2, plot_params=p_color))
pgm.add_node(daft.Node("IP2", '(42.4,71.1)', 1.5, 3.25, aspect=2, scale=1.2, plot_params=p_color))
pgm.add_node(daft.Node("IP3", '(40.0,75.2)', 1.5, 2.5, aspect=2, scale=1.2, plot_params=p_color))
pgm.add_node(daft.Node("u4", r"$u_4$", -0.2, 4.3))
pgm.add_node(daft.Node("u3", r"$u_3$", -1, 3.6))
pgm.add_node(daft.Node("u2", r"$u_2$", -0.6, 2.9))
pgm.add_node(daft.Node("u1", r"$u_1$", -0.4, 2.3))
pgm.add_node(daft.Node("u5", r"$u_5$", 3., 4.1))
pgm.add_node(daft.Node("u6", r"$u_6$", 3., 3.3))
pgm.add_node(daft.Node("u7", r"$u_7$", 3., 2.5))
pgm.add_plate(daft.Plate([0.5, 2.05, 2., 2.5],
        label="IP"))
pgm.add_edge("IP1", "u4")
pgm.add_edge("IP2", "u4")
pgm.add_edge("IP1", "u3")
pgm.add_edge("IP1", "u2")
pgm.add_edge("IP2", "u2")
pgm.add_edge("IP3", "u2")
pgm.add_edge("IP3", "u1")
pgm.add_edge("IP1", "u5")
pgm.add_edge("IP2", "u5")
pgm.add_edge("IP2", "u6")
pgm.add_edge("IP3", "u6")
pgm.add_edge("IP1", "u7")
pgm.add_edge("IP3", "u7")
pgm.render()
pgm.figure.savefig("location-model.png", dpi=300)
```
